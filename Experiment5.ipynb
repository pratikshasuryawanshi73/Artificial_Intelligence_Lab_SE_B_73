```python

import pandas as pd
import numpy as np
data = pd.read_csv("indian_diseases_dataset.csv")
print("First 5 Rows of Dataset:")
print(data.head())
print("\nDataset Information:")
print(data.info())
print("\nNumber of Rows and Columns:")
print(data.shape)
print("\nStatistical Summary:")
print(data.describe())
severe_cases = data[data['severity'] == 'Severe']
print("\nSevere Disease Cases:")
print(severe_cases.head())
old_patients = data[data['age'] > 60]
print("\nPatients Age > 60:")
print(old_patients.head())
high_cost = data[data['treatment_cost_inr'] > 100000]
print("\nHigh Treatment Cost Cases:")
print(high_cost.head())
gender_group = data.groupby("gender").size()
print("\nNumber of Patients by Gender:")
print(gender_group)
disease_group = data.groupby("disease_category").size()
print("\nDisease Category Distribution:")
print(disease_group)
avg_cost = data.groupby("disease_category")["treatment_cost_inr"].mean()
print("\nAverage Treatment Cost by Disease Category:")
print(avg_cost)
avg_stay = data.groupby("severity")["days_hospitalized"].mean()
print("\nAverage Hospital Stay by Severity:")
print(avg_stay)
outcome_count = data["outcome"].value_counts()
print("\nPatient Outcome Distribution:")
print(outcome_count)
avg_recovery = data["recovery_days"].mean()
print("\nAverage Recovery Days:", avg_recovery)
max_cost = data["treatment_cost_inr"].max()
print("Maximum Treatment Cost:", max_cost)
min_cost = data["treatment_cost_inr"].min()
print("Minimum Treatment Cost:", min_cost)

print("\nData Manipulation Completed Successfully.")
```
