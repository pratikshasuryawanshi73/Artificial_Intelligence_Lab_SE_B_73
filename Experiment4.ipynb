```python

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

data = np.random.normal(loc=0, scale=1, size=1000)

print("First 10 Random Numbers:")
print(data[:10])

mean_value = np.mean(data)
median_value = np.median(data)
variance_value = np.var(data)
std_value = np.std(data)

print("\nStatistical Analysis")
print("Mean:", mean_value)
print("Median:", median_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_value)

plt.hist(data, bins=30)
plt.title("Normal Distribution Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()

tosses = 1000

coin = np.random.choice([0,1], size=tosses)

heads = np.sum(coin == 1)
tails = np.sum(coin == 0)

print("\nCoin Toss Simulation (1000 tosses)")
print("Heads:", heads)
print("Tails:", tails)

prob_heads = heads / tosses
prob_tails = tails / tosses

print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```
