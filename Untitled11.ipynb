MEAN, MEDIAN, MODE

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
# Take numbers from user
data = list(map(float, input("Enter numbers separated by space: ").split()))
# Convert list to numpy array
data = np.array(data)
# Calculate mean, median and mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)[0]
# Print results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

VARIENCE

```python
# Compute variance
variance_value = np.var(data)
print(f"Variance: {variance_value:.2f}")
```

STANDARD DEVIATION

```python
# Compute standard deviation
std_dev_value = np.std(data)
print(f"Standard Deviation: {std_dev_value:.2f}")
```

NORAMAL DISTRIBUTATION GRAPH

```python
# Generate x values
x = np.linspace(min(data), max(data), 100)
# Calculate probability density
y = stats.norm.pdf(x, mean, std_dev_value)
# Plot graph
plt.plot(x, y)
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.show()
```

COIN TOSS SIMULATION

```python
# Take number of tosses from user
n = int(input("Enter number of coin tosses: "))
# Simulate coin toss
tosses = np.random.choice(['Head', 'Tail'], size=n)
# Count results
heads = np.sum(tosses == 'Head')
tails = np.sum(tosses == 'Tail')
# Display results
print("Heads:", heads)
print("Tails:", tails)
# Calculate probability
print("Probability of Heads:", heads/n)
print("Probability of Tails:", tails/n)
```
