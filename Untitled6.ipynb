Determinant of the matrix


```python
import numpy as np
A = np.array([
    [1, 2, 3],
    [0, 4, 5],
    [0, 0, 6]
])
det = np.linalg.det(A)
print("Matrix:\n", A)
print(f"Determinant: {det:.2f}")
```

```python
A = [
    [1, 2, 3],
    [0, 4, 5],
    [0, 0, 6]
]
a, b, c = A[0]
d, e, f = A[1]
g, h, i = A[2]
det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)
print("Matrix:")
for row in A:
    print(row)
print("Determinant:", det)
```

Inverse of matrix

```python
A = np.array([
    [1, 2, 3],
    [0, 4, 5],
    [0, 0, 6]
])
inv_A = np.linalg.inv(A)
print("Inverse matrix:")
print(inv_A)
```

```python
print("Inverse matrix:")
print(np.round(inv_A, 2))
```

```python
A = [
    [1, 2, 3],
    [0, 4, 5],
    [0, 0, 6]
]
a, b, c = A[0]
d, e, f = A[1]
g, h, i = A[2]
det = a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)
print("Determinant:", det)
if det == 0:
    print("Inverse does not exist")
else:
    C = [
        [(e*i - f*h), -(d*i - f*g),  (d*h - e*g)],
        [-(b*i - c*h), (a*i - c*g), -(a*h - b*g)],
        [(b*f - c*e), -(a*f - c*d),  (a*e - b*d)]
    ]
    adj = [
        [C[0][0], C[1][0], C[2][0]],
        [C[0][1], C[1][1], C[2][1]],
        [C[0][2], C[1][2], C[2][2]]
    ]
    inv = [
        [adj[0][0]/det, adj[0][1]/det, adj[0][2]/det],
        [adj[1][0]/det, adj[1][1]/det, adj[1][2]/det],
        [adj[2][0]/det, adj[2][1]/det, adj[2][2]/det]
    ]
    print("Inverse matrix:")
    for row in inv:
        print(row)
```

```python
print(np.round(inv_A, 2))
```

Linear equation solving

```python
import numpy as np
A = np.array([
    [2, 3],
    [4, 5]
])
B = np.array([10, 20])
solution = np.linalg.solve(A, B)
x = solution[0]
y = solution[1]
print("x =", x)
print("y =", y)
```

```python
a1, b1, c1 = 2, 3, 10
a2, b2, c2 = 4, 5, 20
D  = a1*b2 - a2*b1
Dx = c1*b2 - c2*b1
Dy = a1*c2 - a2*c1
if D == 0:
    print("No unique solution")
else:
    x = Dx / D
    y = Dy / D
    print("x =", x)
    print("y =", y)
```
