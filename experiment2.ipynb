BFS AND DFS


```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}


def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)

    print("BFS Traversal:", end=" ")

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    print()


def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
        print("DFS Traversal:", end=" ")

    visited.add(start)
    print(start, end=" ")

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)



if __name__ == "__main__":
    bfs(graph, 'A')
    dfs(graph, 'A')

```

A* algorithm

```python
import heapq
import math


graph = {
    'A': {'B': 6, 'F': 3},
    'B': {'C': 3, 'D': 2},
    'C': {'D': 1, 'E': 5},
    'D': {'E': 8},
    'E': {'G': 5},
    'F': {'G': 1, 'E': 7},
    'G': {}
}


node_positions = {
    'A': (0, 0),
    'B': (2, 3),
    'C': (5, 4),
    'D': (4, 1),
    'E': (7, 2),
    'F': (1, -2),
    'G': (8, -1)
}

def heuristic(node, goal, positions):

    x1, y1 = positions[node]
    x2, y2 = positions[goal]
    return math.sqrt((x1 - x2)**2 + (y1 - y2)**2)

def a_star(graph, start, goal, positions):


    closed_set = set()


    open_set = [(heuristic(start, goal, positions), start)]


    came_from = {}


    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0


    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic(start, goal, positions)

    while open_set:

        current_f, current_node = heapq.heappop(open_set)

        if current_node == goal:
            return reconstruct_path(came_from, current_node)

        closed_set.add(current_node)

        for neighbor, weight in graph[current_node].items():
            if neighbor in closed_set:
                continue


            tentative_g_score = g_score[current_node] + weight

            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current_node
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + heuristic(neighbor, goal, positions)
                # Add neighbor to open_set if it's not already there with a better f_score
                heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return None

def reconstruct_path(came_from, current):
    """Reconstructs the path from the came_from dictionary."""
    path = [current]
    while current in came_from:
        current = came_from[current]
        path.append(current)
    return path[::-1]

# Example Usage:
start_node = 'A'
goal_node = 'G'

path = a_star(graph, start_node, goal_node, node_positions)

if path:
    print(f"Shortest path from {start_node} to {goal_node}: {path}")
    # Calculate total cost of the path
    total_cost = 0
    for i in range(len(path) - 1):
        u, v = path[i], path[i+1]
        total_cost += graph[u][v]
    print(f"Total cost: {total_cost}")
else:
    print(f"No path found from {start_node} to {goal_node}")
```

Minmax algorithm


```python
import math



game_tree_minimax = {
    'Root': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['E', 'F'],
    'C': [1, 2],  # C's children are direct scores (terminal states)
    'D': [3, 4],  # D's children are direct scores
    'E': [5, 6],  # E's children are direct scores
    'F': [7, 8]   # F's children are direct scores
}

def minimax_algorithm(node_key, maximizing_player):
    """Implements the Minimax algorithm for a given game tree.

    Args:
        node_key (str): The current node in the game tree (key in game_tree_minimax).
        maximizing_player (bool): True if it's the maximizing player's turn, False for minimizing.

    Returns:
        int: The optimal value for the current node.
    """
    # If the current node_key maps to a list of integers, it's a terminal state
    if isinstance(game_tree_minimax.get(node_key), list) and all(isinstance(val, int) for val in game_tree_minimax[node_key]):
        if maximizing_player:
            return max(game_tree_minimax[node_key])
        else:
            return min(game_tree_minimax[node_key])

    # If it's an internal node, recurse
    if maximizing_player:
        best_value = -math.inf
        for child_key in game_tree_minimax[node_key]:
            value = minimax_algorithm(child_key, False) # Next player is minimizing
            best_value = max(best_value, value)
        return best_value
    else: # Minimizing player
        best_value = math.inf
        for child_key in game_tree_minimax[node_key]:
            value = minimax_algorithm(child_key, True) # Next player is maximizing
            best_value = min(best_value, value)
        return best_value


optimal_value_root = minimax_algorithm('Root', True)
print(f"The optimal value for the Maximizing player starting at Root is: {optimal_value_root}")


optimal_value_A_min = minimax_algorithm('A', False)
print(f"The optimal value for the Minimizing player starting at A is: {optimal_value_A_min}")


optimal_value_B_min = minimax_algorithm('B', False)
print(f"The optimal value for the Minimizing player starting at B is: {optimal_value_B_min}")
```
